# EscapeRoom_MAPOCA

```python
import os
import platform

# Global Setting
cur_dir = os.getcwd()
env_dir = os.path.abspath(os.path.join(cur_dir, "..", "Unity6000_Envs"))
output_dir = os.path.abspath(os.path.join(cur_dir, "temp", "mlagents_learn_output"))
```

```python
# Unity Enviroment
game = "EscapeRoom"
os_name = platform.system()

if os_name == 'Linux':
    env_name = os.path.join(env_dir, f"{game}_{os_name}.x86_64")
elif os_name == 'Darwin':
    env_name = os.path.join(env_dir, f"{game}_{os_name}.app")
env_fp = os.path.join(env_dir, env_name)
print(env_fp)
baseport = 1967
```

## Training PPO


```python
config_ppo_fp = os.path.join(cur_dir, "config", "EscapeRoom_ppo.yaml")
run_ppo_id = "EscapeRoom_PPO"
print(config_ppo_fp)
print(run_ppo_id)

!mlagents-learn $config_ppo_fp \
               --env=$env_fp \
               --results-dir=$output_dir \
               --run-id=$run_ppo_id --base-port=$baseport
```

## Training SAC


```python
config_sac_fp = os.path.join(cur_dir, "config", "EscapeRoom_sac.yaml")
run_sac_id = "EscapeRoom_SAC"
print(config_sac_fp)
print(run_sac_id)

!mlagents-learn $config_sac_fp \
               --env=$env_fp \
               --results-dir=$output_dir \
               --run-id=$run_sac_id --base-port=$baseport
```
